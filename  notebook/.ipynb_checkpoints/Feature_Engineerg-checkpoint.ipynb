```python
num_features=mydf_train.select_dtypes([float,int]).columns
cat_features=mydf_train.select_dtypes(object).columns
```

```python
##Importing the important libraries for the prediction
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
num_pipeline=Pipeline(steps=[(
    "imputer",SimpleImputer(strategy="median")),
    ("sacaler",StandardScaler())
])
```

```python
cat_pipeline=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("scaler",OneHotEncoder())
])
```

```python
num_pipeline
```

```python
cat_pipeline
```

```python
Preprocessor=ColumnTransformer([
    ("num_pipeline",num_pipeline,num_features),
    ("cat_pipeline",cat_pipeline,cat_features)
])
```

```python
Preprocessor
```

```python
Preprocessor=Preprocessor.fit(mydf_train)
```

```python
Preprocessor
```

```python
import pickle
```

```python
with open("Preprocessor.pkl","wb") as f:
  pickle.dump(Preprocessor,f)
```
