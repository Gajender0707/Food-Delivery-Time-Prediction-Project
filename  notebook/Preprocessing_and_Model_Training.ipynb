```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
y=training_df["Time_taken"]
y
```

```python
x=training_df.drop("Time_taken",axis=1)
```

```python
x.head()
```

```python
testing_df.shape
```

```python
testing_df.drop("Time_taken",axis=1,inplace=True)
```

```python
testing_df.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
num_features=x_train.select_dtypes(include=["int64","float64"]).columns
cat_features=x_train.select_dtypes(include=["object"]).columns
```

```python
num_features
```

```python
cat_features
```

```python
num_pipeline=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="median")),
    ("scaler",StandardScaler())
])
```

```python
cat_pipeline=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("onehot",OneHotEncoder())
])
```

```python
Preprocessor=ColumnTransformer([
    ("num_pipeline",num_pipeline,num_features),
    ("cat_pipeline",cat_pipeline,cat_features)
])
```

```python
Preprocessor
```

```python
Preprocessor=Preprocessor.fit(x_train)
```

```python
Preprocessor
```

```python
x_train=Preprocessor.transform(x_train)
```

```python
x_train=x_train.toarray()
```

```python
models={
    "LinearRegression":LinearRegression(),
    "RandomForestRegressor":RandomForestRegressor(),
    "DecisionTreeRegressor":DecisionTreeRegressor(),
    "XGBRegressor":XGBRegressor(),
    "LGBMRegressor":LGBMRegressor(),
    "Lasso":Lasso(),
    "Ridge":Ridge()
}

```

```python
list(models.values())[0]
```

```python
def report(models,x_train,y_train,x_test,y_test):
  for i in range(len(models)):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    score=model.score(x_test,y_test)
    print(list(models.keys())[i],score)
```

```python
report(models,x_train,y_train,x_test,y_test)
```
